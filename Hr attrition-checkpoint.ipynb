Class imbalance issue

```python
from imblearn.over_sampling import SMOTE
```

```python
sm=SMOTE()
```

```python
x,y=sm.fit_resample(x,y)
```

```python
y.value_counts()
```

Model building

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

###### Finding best random state

```python
from sklearn.tree import DecisionTreeClassifier
maxAccu=0
maxRs=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    mod=DecisionTreeClassifier()
    mod.fit(x_train,y_train)
    pred=mod.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRs=i
print("Best Accuracy is:",maxAccu,"on Random state:",maxRs)
```

best accuracy is 84% and random state is 36

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=maxRs)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
DTC=DecisionTreeClassifier()
DTC.fit(x_train,y_train)
pred=DTC.predict(x_test)
acc=classification_report(y_test,pred)
print(acc)
```

Decision Tree gives 84% accuracy

```python
RFC=RandomForestClassifier()
RFC.fit(x_train,y_train)
pred=RFC.predict(x_test)
acc=classification_report(y_test,pred)
print(acc)
```

Random forest gives 91% accuracy

```python
sv=SVC()
sv.fit(x_train,y_train)
pred=sv.predict(x_test)
acc=classification_report(y_test,pred)
print(acc)
```

Support vector classifier gives 64% accuracy

Upon reviewing the above outputs the Random forest classifier gives 91% accuracy.So to choose the best model we will use cross validation Score.

```python
from sklearn.model_selection import cross_val_score
```

```python
print(cross_val_score(DTC,x,y,cv=5).mean())
```

Decision Tree gives 65.3%

```python
print(cross_val_score(RFC,x,y,cv=5).mean())
```

Random fores gives 77%

So our random forest classifier gives the less difference so we will consider this a best model
